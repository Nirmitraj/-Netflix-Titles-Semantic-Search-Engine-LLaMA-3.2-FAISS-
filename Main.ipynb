You’re combining all the relevant info into one descriptive text string so that:

-> It can be used as input to a language model (like for embeddings).
-> The structure makes it easier for models to understand relationships between fields.
-> It's a more human-readable version for debugging or displaying info.

```python
def create_textual_representation(row):
    textual_representation = f"""Type: {row['type']},
    Title: {row['title']},
    Director: {row['director']},
    Cast: {row['cast']},
    Released: {row['release_year']},
    Genres: {row['listed_in']},

    Decription: {row['description']} """

    return textual_representation
```

```python
df['textual_representation'] = df.apply(create_textual_representation, axis=1)
```

```python
embedding = np.array([res.json()['embedding']], dtype = 'float32' )
#Get the embedding and convert to NumPy
D, I = index.search(embedding, 5) #Search the FAISS index
```

```python
best_matches = np.array(df['textual_representation'])[I.flatten()]
```
